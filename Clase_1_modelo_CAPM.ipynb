```python
precios_amazon = dat.history(period='16y',interval='1mo',end='2024-12-31')
type(precios_amazon)
cierre_amazon = precios_amazon['Close']

precios_dow = dat1.history(period='16y',interval='1mo',end='2024-12-31')
cierre_dow = precios_dow['Close']
```

```python
#crear funcion para calcular la rentabilidad de cada acción
def calcula_rentabilidad_promedio(precios):
   historica = precios.pct_change()
   return historica
```

```python
rentabilidad_empresa = calcula_rentabilidad_promedio(cierre_amazon)
rent_empresa_df = pd.DataFrame(rentabilidad_empresa,columns=['Close'])
#rent_empresa_df =rent_empresa_df.dropna()
```

```python
rent_empresa_df
```

```python
print('La rentabilidad promedio de la empresa es: ',rent_empresa_df.mean()*100,'%')
```

```python
rentabilidad_mercado = calcula_rentabilidad_promedio(cierre_dow)
rent_mercado_df = pd.DataFrame(rentabilidad_mercado,columns=['Close'])
rent_mercado_df
```

```python
print('La rentabilidad promedio del mercado es: ',rent_mercado_df.mean()*100,'%')
```

```python
#Para obtener beta utilizamos la pendiente o regresión lineal entre la serie de rentabilidad de la empresa y el mercado

y_conocido = rent_empresa_df.dropna()
x_conocido = rent_mercado_df.dropna()

modelo = LinearRegression()

modelo.fit(x_conocido,y_conocido)

pendiente = modelo.coef_[0]
```

```python
pendiente
```

```python
total = pendiente + rentabilidad_empresa.mean() + rentabilidad_mercado.mean()
```

```python
print('El retorno esperado para el inversionista es:', (total))
```
